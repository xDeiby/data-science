# Plots y Visualización de los datos

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("C:/Users/deiby/Documents/curso-ml/datasets/customer-churn-model/Customer Churn Model.txt")
dataset.head()
```

* Función magica, utilizada para que la representacion del grafico se muestre en linea

```python
%matplotlib inline
```

## Scatter plot (Nube de puntos o dispersión)

```python
dataset.plot(kind="scatter", x = "Day Mins", y = "Day Charge")
```

```python
dataset.plot(kind="scatter", x = "Night Mins", y = "Night Charge")
```

* Aquí se crea una matriz de graficos, a la cual podemos asginar graficos

```python
figure, axs = plt.subplots(2,2, sharey= True, sharex= True)
dataset.plot(kind="scatter", x = "Day Mins", y = "Day Charge", ax=axs[0][0])
dataset.plot(kind="scatter", x = "Night Mins", y = "Night Charge", ax=axs[0][1])
dataset.plot(kind="scatter", x = "Day Calls", y = "Day Charge", ax=axs[1][0])
dataset.plot(kind="scatter", x = "Night Calls", y = "Night Charge", ax=axs[1][1])
```

## Gráfico Histograma
* *bins* le indica la cantidad de divisiones, o cajas. Lo cual podemos indicar mediante un numero o una sequencia
* La regla mas utilizada es la de **sturges** la cual segun nuestro tamaño de muestra aplica la formula *1+ log(M)*

```python
# Ceil aproxima hacia arriba
k = int(np.ceil(np.log2(3333)))
# k o bien "sturges"
plt.hist(dataset["Day Calls"], bins = "sturges")
plt.xlabel("Número de llamadas el día")
plt.ylabel("Frecuencia")
plt.title("Histograma de número de llamadas al día")
```

## Boxplot, diagramas de cajas y bigotes

```python
plt.boxplot(dataset["Day Calls"])
plt.ylabel("Número de llamadas diarias")
plt.title("Boxplot de las llamas diarias")
```
