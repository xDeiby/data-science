- **sep**: parametro que indica como seran separados los datos (Por defecto: ,)
- **dtype**: parametro mediante el cual podemos transformar una columna de su formato origen a el que desiemos (Por defecto: None)

- **skip_blank_lines**: se saltara las lineas en blanco en el dataset (Por defecto: False)

# Cargar datos mediante Open

- Debido a que ***open_csv*** al cargar el archivo, debe leer el dataset completo, puede ser ineficiente si trabajos datasets de un gran tamaño (Ej: 10 Gigas), por lo tanto es mucho mas conveniente utilizar ***open***
- Lo que hace ***open*** lee linea por linea el dataset y una vez es leida la linea con un ciclo for, se elimina del cache
- Abre el los dataset en modo lectura, no almacena los datos, lo cual es conveniente para guardar fragmentos del dataset

## Formas de abrir el archivo
- **r** - lectura
- **w** - escritura
- **rw** - lectura / escritura

> Si lo que se quiere es agregar informacion al fichero, lo recomendable es abrirlo en **a**, ya que esto posicionara el cursor al final del fichero, mientras que **w** sobrescribira

## **strip()** -> elimina los espacios en blanco de un string (La sangria)

```python
dataset3 = open(path2, "r")
cols = dataset3.__next__().strip().split(",")
n_cols = len(cols)

cols
```

```python
n_rows = 0
main_dict = {}
for col in cols:
    main_dict[col] = []

main_dict
```

```python
for line in dataset3:
    values = line.strip().split(",")
    for i in range(n_cols):
        main_dict[cols[i]].append(values[i])
    n_rows += 1;

print("El dataset tiene %d columnas y %d filas" % (n_cols, n_rows))
```

## Una vez ya se tiene toda la informacion en el **diccionario**, podemos asignar esta informacion en un **DataFrame**

```python
df3 = pd.DataFrame(main_dict)

df3.head()
```

## Lectura y escritura de una archivo
- Leemos un dataset, y abrimos otro dataset en modo escritura. Finalmente escribimos linea por linea el primer dataset en el segundo, con el delimitador definido
- **with** sentencia con la misma funcion que un **try-finally**. Es algo que se va a realizar simplemente en un bloque, es decir si por ejemplo abrimos el archivo, al finalizar el bloque, se cerraran automaticamente, ya que nos salimos del **contexto**

```python
dataset4 = "customer-churn-model\Customer Churn Model.txt"
dataset5 = "customer-churn-model\Tab Customer Churn Model.txt"
path4 = os.path.join(basepath, dataset4)
path5 = os.path.join(basepath, dataset5)

with open(path4, "r") as infile:
    with open(path5, "w") as outfile:
        for line in infile:
            line = line.strip().split(",")
            outfile.write("\t".join(line)+"\n")
```

```python
df4 = pd.read_csv(path5, sep="\t");
df4.head()
```

## Lectura mediante urllib

```python
import urllib3
import csv

http = urllib3.PoolManager()
result = http.request("GET", "https://raw.githubusercontent.com/joanby/python-ml-course/Collab---v-3.8/datasets/customer-churn-model/Tab%20Customer%20Churn%20Model.txt")
result.status
# result.data
```

- Transformamos el **result.data** de tipo ***byte*** -> ***str*** para que el archivo se pueda leer.
- Ya teniendo solo un string, particionamos el str por saltos de linea, es decir **\n**. Creando un array, donde cada indice es la linea del archivo
- Luego mediante el metodo ***csv.reader** creamos una matriz de 1xcantidad_elementos, es decir creamos las filas, pero no las columnas

```python
response = result.data
decoded = response.decode("utf-8")
lines_data = decoded.split("\n")
file_output = "../datasets/write_csv/NewDataset.txt"
cr = csv.reader(lines_data)

with open(file_output, "w") as new_file:
    for row in cr:
        if(len(row) > 0):
            new_file.write(row[0])
            new_file.write("\n")

result = pd.read_csv(file_output, sep="\t")
result.columns.values
```

```python
def downloadToFormat(path, filename, url, sep = ",", delim = "\n", format="csv", encoding="utf-8", is_header = True):
    # Se forma la url
    destiny_path = os.path.join(path, filename) + "." + format
    
    # Se obtiene el dataset
    http = urllib3.PoolManager()
    response = http.request("GET", url)

    # Se transforma a string
    data = response.data.decode(encoding)

    # Array de filas
    lines = data.split(delim)

    # Numero de Filas
    n_rows = len(lines)

    # Si el dataset tiene header, se utiliza, sino se crea uno
    if(is_header):
        header = lines[0].strip().split(sep)
        lines.pop(0)
    else:
        header = []

    # Se guardan las cabeceras en un diccionario
    dataset_dict = {}
    for col in header:
        dataset_dict[col] = []

    n_cols = len(dataset_dict)
    
    for line in lines:
        line = line.strip().split(sep)
        for index in range(n_cols):
            dataset_dict[header[index]].append(line[index].strip())
    

    df = pd.DataFrame(dataset_dict)

    if(format == "csv"):
        df.to_csv(destiny_path, index= False)
    elif(format == "json"):
        df.to_json(destiny_path)
    elif(format == "xls"):
        # Its deprecated
        df.to_excel(destiny_path, index= False)


    return df






new_df = downloadToFormat(
    path="C:/Users/deiby/Documents/curso-ml/datasets/write_csv",
    filename= "outuputfile",
    url= "https://raw.githubusercontent.com/joanby/python-ml-course/master/datasets/athletes/Medals.csv",
    encoding= "cp1252",
    format= "xls"
)

new_df.head()
```
