# GeneraciÃ³n de nÃºmeros pseudo aleatorios

```python
import numpy as np
```

```python
np.random.randint(0,100)
```

```python
# La forma mas clÃ¡sica de generar nÃºmeros aleatorios, es entre 0 y 1
np.random.random()
```

```python
# Una funciÃ³n de nÃºmero pseudo aleatorios
def randint_list(n, a, b):
    x = []
    for i in range(n):
        x.append(np.random.randint(a,b))

    return x
```

```python
randint_list(20, 0,100)
```

```python
# Sin embargo, esto se puede hacer mediante el modulo random
import random

# El tercer parametro indica los multiplo de que numero queremos
for i in range(10):
    print(random.randrange(1,100, 7))
```
