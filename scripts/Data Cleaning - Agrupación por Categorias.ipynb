# Agrupación por Categorías

```python
import numpy as np
import pandas as pd
```

```python
gender = ["Female", "Male"]
incomes = ["Poor", "Middle Class", "Rich"]
n_rows = 500

# Generamos una muestra de 500 datos de genero e ingresos, segun los arreglos descritos ariiba
gender_sample = np.random.choice(gender, size = n_rows)
incomes_sample = np.random.choice(incomes, size = n_rows)
```

```python
gender_sample[400:]
```

```python
# Generamos 500 alturas, con una distribución normal N(160, 30)
heights = 160 + 30 * np.random.randn(n_rows)

# Generamos 500 pesos, con una distribución normal N(65, 25)
weights = 65 + 25 * np.random.randn(n_rows)

# Generamos 500 edades, con una distribución normal N(30, 12)
ages = 30 + 12 * np.random.randn(n_rows)

# Generamos 500 ingresos, con una distribución normal N(230000, 50000)
incomes = 230000 + 50000 * np.random.randn(n_rows)
```

```python
# Se generá el dataset
data = pd.DataFrame(
    {
        "Gender": gender_sample,
        "Economic Status": incomes_sample,
        "Income": incomes,
        "Weight": weights,
        "Height": heights,
        "Age": ages.round(decimals= 0)

    }
)
data.max()
```

```python
group_gender = data.groupby("Gender")
group_gender.groups
```

```python
for name, group in group_gender:
    print(name)
    print(group) 
```

## Filtrado de dato

```python
# Double group es una SerieGroupBy
double_group["Age"].sum()
```

```python
# Al filtrar por suma, toma cada grupo y suma todos sus elementos y los filtra, devolviendo una Serie con todo el cunjunto de numero que contienen los grupos que cumplen la condicion
double_group["Age"].filter(lambda a: a.sum() > 2600)

# Seleccionar una categoria multiple
# double_group.get_group(("Male", "Middle Class"))
```

## Transformación de Variables

```python
# Ademas de tificar los valores, transforman los datos de un DataFrameGroupBy a un DataFrame
zscore = lambda x:( x - x.mean() )/ x.std()
z_group = double_group.transform(zscore)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.hist(z_group["Age"])
```

```python
# Por ejemplo si quere remplazar los valores vacios por el promedio
fill_na_mean = lambda x: x.fillna(x.mean())

# z_group.transform(fill_na_mean)
double_group.transform(fill_na_mean)
```
