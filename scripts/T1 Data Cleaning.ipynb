## Borrado de valores que faltan
* El motodo de pandas **dropna()** se utiliza para borrar la fila correspondiente a los valos nulos (axis = 0) o bien la columna (axis = 1)
* **how** indica en que caso queremos que pase, en este caso *all* indica que si la fila se compone solo de valores nulos, sera eliminada. Si de define *any*, si alguna de las columnas contiene nulo, se elimina toda la fila

```python
dataset.dropna(axis = 0, how="all")
```

```python
dataset.dropna(axis=0, how="any")
```

## Computo de los valores faltantes
* Es decir, remplazar los valores nulos por otro string, 0 o el promedio, etc
* **fillna()** remplaza los valores nulos por el parametro que le indiquemos

```python
# dataset.fillna(0)
dataset.fillna("Desconocido")
```

* Tambien podemos decidir a que columna vamos a aplicar

```python
dataset["body"] = dataset["body"].fillna(0)
dataset["home.dest"] = dataset["home.dest"].fillna("Desconocido")

dataset.tail(10)
```

```python
pd.isnull(dataset["age"]).values.sum()
age_mean = dataset["age"].mean()
dataset["age"].fillna(age_mean)
```

* Otro metodo utilizado, es remplazar los valores, por el valor que tienen mas cerca de si mismo hacia delante (*ffill*) o atras. (Valores no nulos)
* *inplace* utilizado para que permanezcan los cambios en el dataset

```python
dataset["age"].fillna(method="ffill")
dataset["age"].fillna(method="backfill", inplace=True)
```

## Función de creación de variable dummies

```python
def createDummies(dataset, col):
    dummy_col = pd.get_dummies(dataset[col], prefix=col)
    new_dataset = dataset.drop(col, axis=1)
    new_dataset = pd.concat([new_dataset, dummy_col], axis=1)

    return new_dataset
```

```python
createDummies(dataset, "sex")
```
